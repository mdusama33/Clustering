```python
crime=pd.read_csv('D:\\Data Science\\Assignments\\Assignment 7 (Clustering)\\crime_data.csv')
crime
```

```python
# Normalize Function
def minmax_func(i):
    x=(i-i.min())/(i.max()-i.min())
    return(x)
```

```python
df_norm=minmax_func(crime.iloc[:,1:])
df_norm
```

```python
# Create dendogram
dendogram=sch.dendrogram(sch.linkage(df_norm,method='average'))
```

```python
# Create Clusters
hc=AgglomerativeClustering(n_clusters=3,affinity='euclidean',linkage='complete')
hc
```

```python
y_hc=hc.fit_predict(df_norm)
print(y_hc)
Clusters=pd.DataFrame(y_hc,columns=['clusters'])
```

```python
crime['h_clusterid']=hc.labels_
crime
```

# Kmeans

```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
from sklearn.cluster import KMeans
```

```python
crime1=pd.read_csv('D:\\Data Science\\Assignments\\Assignment 7 (Clustering)\\crime_data.csv')
crime1
```

```python
def norm_func(i):
    x=(i-i.mean())/(i.max()-i.min())
    return(x)
```

```python
x=norm_func(crime1.iloc[:,1:])
x
```

```python
# Normalize function
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
scaled_crime1_df=scaler.fit_transform(crime1.iloc[:,1:])
```

```python
wcss=[]
for i in range (1,11):
    Kmeans=KMeans(n_clusters=i,random_state=0)
    Kmeans.fit(scaled_crime1_df)
    wcss.append(Kmeans.inertia_)
    
plt.plot(range(1,11),wcss)
plt.title('Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('wcss')
plt.show()
```

```python
#build cluster algorithm
from sklearn.cluster import KMeans
clusters_new = KMeans(4,random_state=0)
clusters_new.fit(scaled_crime1_df)
```

```python
clusters_new.labels_
```

```python
#Assign clusters to the dataset
crime1['clusterid_new']=clusters_new.labels_
crime1
```

```python
#These are standardized values
clusters_new.cluster_centers_
```

```python
crime1.groupby('clusterid_new').agg(['mean']).reset_index()
```

```python
crime1.columns
```

```python
crime1['clusterid_new'].value_counts()
```
